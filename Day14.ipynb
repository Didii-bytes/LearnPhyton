```python
countries = ['Estonia', 'Finland', 'Sweden', 'Denmark', 'Norway', 'Iceland']

def change_to_upper(x):
    return x.upper()

contries_upper_cased = map(change_to_upper,countries)
print(list(contries_upper_cased))
```

4. Use filter to filter out countries containing 'land'.

```python
countries = ['Estonia', 'Finland', 'Sweden', 'Denmark', 'Norway', 'Iceland']

def filter_land(x):
    if 'land' in x:
        return True
    else:
        return False
    
filter_countries=filter(filter_land,countries)
print(list(filter_countries))
```

5. Use filter to filter out countries having exactly six characters.

```python
def length_countries(x):
    if len(x) == 6:
        return True
    else:
        return False

filter_length_countries=filter(length_countries,countries)
print(list(filter_length_countries))
```

9. Declare a function called get_string_lists which takes a list as a parameter and then returns a list containing only string items.

11. Use reduce to concatenate all the countries and to produce this sentence: Estonia, Finland, Sweden, Denmark, Norway, and Iceland are north European countries

```python
from functools import reduce

# Use reduce to concatenate the country names with commas and "and"
sentence = reduce(lambda x, y: f"{x}, {y}" 
                  if y != countries[-1] 
                  else f"{x}, and {y}", countries)

# Append the final part of the sentence
sentence += " are north European countries."

print(sentence)
```

12. Declare a function called categorize_countries that returns a list of countries with some common pattern (you can find the countries list in this repository as countries.js(eg 'land', 'ia', 'island', 'stan')).

```python
from countries import countries

def categorize_countries(countries_list,pattern):
    return [country for country in countries_list if pattern in country]

land_countries = categorize_countries(countries, "land")
ia_countries = categorize_countries(countries, "ia")
island_countries = categorize_countries(countries, "island")
stan_countries = categorize_countries(countries, "stan")

print("Countries with 'land':", land_countries)
print("Countries with 'ia':", ia_countries)
print("Countries with 'island':", island_countries)
print("Countries with 'stan':", stan_countries)
```

13. Create a function returning a dictionary, where keys stand for starting letters of countries and values are the number of country names starting with that letter.

```python
from countries import countries

def count_countries_by_letter(countries):
    country_count = {}
    
    for country in countries:
        first_letter = country[0].upper()  # Get the first letter and ensure uppercase
        country_count[first_letter] = country_count.get(first_letter, 0) + 1
    
    return country_count

# Call the function
result = count_countries_by_letter(countries)

print(result)
```
